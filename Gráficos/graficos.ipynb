```python
def cálculo_de_cuota_ganadora_dosopc(dataframe,n_columna): 
    lista_cuotas = []
    for i in range(len(dataframe)):
        if dataframe.iloc[i,n_columna]=='L':
            cuota=dataframe.iloc[i,7]
            lista_cuotas.append(round(cuota,2))
        elif dataframe.iloc[i,n_columna]=='E':
            cuota=dataframe.iloc[i,8]
            lista_cuotas.append(round(cuota,2))
        elif dataframe.iloc[i,n_columna]=='V':
            cuota=dataframe.iloc[i,9]
            lista_cuotas.append(round(cuota,2))
        elif dataframe.iloc[i,n_columna]=='LE':
            cuota=dataframe.iloc[i,10]
            lista_cuotas.append(round(cuota,2))
        elif dataframe.iloc[i,n_columna]=='LV':
            cuota=dataframe.iloc[i,11]
            lista_cuotas.append(round(cuota,2))
        elif dataframe.iloc[i,n_columna]=='VE':
            cuota=dataframe.iloc[i,12]
            lista_cuotas.append(round(cuota,2))
        else:
            cuota=0
            lista_cuotas.append(round(cuota,2))

    return lista_cuotas
```

```python
def funcion_premio(df,lista_j, nombre_columna):
    listas_cuotas_dosopciones = []
    for i in range(len(lista_j)):
        cuota = round(np.prod(df[df['Label']==lista_j[i]][nombre_columna]),2)
        listas_cuotas_dosopciones.append(cuota)
    return listas_cuotas_dosopciones
```
