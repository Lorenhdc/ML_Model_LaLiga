```python
import pandas as pd
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score



```

```python
dataset = pd.read_csv('../CSVs/Dataset_practica_completo.csv')
dataset.columns
```

# DATASET COMPLETO

```python
dataset
```

```python
dataset.rename(columns={"Target": "Resultado"}, inplace=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
X = dataset.iloc[:,1:-1]
y = dataset[['Resultado']]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

data_train, data_test, target_train, target_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
pipe = Pipeline(steps = [
    #("scaler", StandardScaler()),
    ("classifier", RandomForestClassifier())
])
pipe
```

```python
random_forest_params = {
    'classifier': [RandomForestClassifier()],
    'classifier__min_samples_split':[2,5,10,15,25],
    'classifier__max_depth':[5,10,15,25,30]
    }
```

```python
gsearch = GridSearchCV(pipe,random_forest_params,cv=10, scoring='accuracy')
gsearch.fit(data_train,target_train)
```

```python
gsearch.best_score_
```

```python
gsearch.best_params_
```

```python
scores = cross_val_score(gsearch.best_estimator_, data_train, target_train, cv=10)
scores
```
