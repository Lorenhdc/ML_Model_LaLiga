```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('Temporadas2000-2021.csv')
```

```python
Target = []
for a in range(len(dataset)): 
    if dataset.loc[a,'Goles Local'] > dataset.loc[a,'Goles Visitante']:
        Target.append('L')
    elif dataset.loc[a,'Goles Local'] == dataset.loc[a,'Goles Visitante']:
        Target.append('E')
    else: 
        Target.append('V')
```

```python
dataset.insert(9,'Target',Target,True)
```

# Penaltis

```python
df_penaltis_local = pd.pivot_table(dataset, columns ='Penaltis Local',index='Temporada',aggfunc='count')['Amarillas Local']
df_penaltis_local.fillna(0, inplace=True)
df_penaltis_local['Total Local'] = df_penaltis_local[1] + df_penaltis_local[2] + df_penaltis_local[3]
df_penaltis_local
```

```python
df_penaltis_visitante = pd.pivot_table(dataset, columns ='Penaltis Visitante',index='Temporada',aggfunc='count')['Amarillas Local']
df_penaltis_visitante.fillna(0, inplace=True)
df_penaltis_visitante['Total Visitante'] = df_penaltis_visitante[1] + df_penaltis_visitante[2]
df_penaltis_visitante
```

```python
# Comparativa penaltis local vs Visitante
lista_visitante_pen = list(df_penaltis_visitante['Total Visitante'])
lista_local_pen = list(df_penaltis_local['Total Local'])
```

```python
tabla_comparativa_pen = pd.DataFrame(lista_local_pen, index=df_penaltis_visitante.index, columns=['Local'])
tabla_comparativa_pen['Visitante'] = lista_visitante_pen
tabla_comparativa_pen ['Total'] = tabla_comparativa_pen['Local'] + tabla_comparativa_pen['Visitante']
tabla_comparativa_pen ['Diff'] = tabla_comparativa_pen['Local'] - tabla_comparativa_pen['Visitante']
tabla_comparativa_pen ['%L'] = round(tabla_comparativa_pen['Local']/ (tabla_comparativa_pen['Local'] + tabla_comparativa_pen['Visitante']),2)
tabla_comparativa_pen ['%V'] = round(tabla_comparativa_pen['Visitante']/ (tabla_comparativa_pen['Local'] + tabla_comparativa_pen['Visitante']),2)
tabla_comparativa_pen
```
